```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Sequential
import sklearn.model_selection as ms
```

```python
df = df.rename(columns={"AveragePrice": "target"})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['month'] = df.Date.dt.month
df['day'] = df.Date.dt.day
```

```python
df = df.drop("Date", axis = 1)
df = df.drop("Lp", axis = 1)
```

```python
stacked =df[['type']].stack()
df[['type']]=pd.Series(stacked.factorize()[0],index=stacked.index).unstack()
stacked =df[['region']].stack()
df[['region']]=pd.Series(stacked.factorize()[0],index=stacked.index).unstack()
X = df.drop("target",axis =1)
```

```python
abs(df.corr()).sort_values(by="target")
```

```python
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size=0.1, random_state=12)
```

```python
y_test.hist()
```

```python
y_train.hist()
```

```python
y.hist()
```

```python
y_test.value_counts()
```

```python
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(1024, activation="relu"),  # numOfNeurons,activationFunction
    Dense(1024, activation="relu"),
    Dense(512, activation="relu"),
    Dense(512, activation="relu"),
    Dense(250, activation="relu"),
    Dense(250, activation="relu"),
    Dense(125, activation="relu"),
    Dense(125, activation="relu"),
    Dense(50, activation="relu"),
    Dense(25, activation="relu"),    
    Dense(1)
])
# network parameters
model.compile(
    loss="mse",
    optimizer="adam",
    metrics=["MAPE"])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```

```python
plt.plot(history.history["val_MAPE"])

np.min(history.history["val_MAPE"])
```
